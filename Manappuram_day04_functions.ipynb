** Simple Lambda Function (Squaring a Number):**

```python
square = lambda x: x * x
print(square(5))
# Output: 25
```

```python
add_numbers = lambda a, b: a + b
print(add_numbers(10, 7))
# Output: 17
```

```python
numbers = [1, 2, 3, 4, 5]
doubled_numbers = list(map(lambda x: x * 2, numbers))
print(doubled_numbers)
# Output: [2, 4, 6, 8, 10]
```

```python
x1= [6,9,11]
y1 = [13,17,1]

x2 = [7,10,12]
y2=[7,16,8]
plt.plot(x1,y1)
plt.plot(x2,y2)
```

Explore the Data

```python
import pandas as pd
filepath="/content/medals_by_country_2016.csv"
data = pd.read_csv(filepath)
data = data.rename(columns={'Unnamed: 0': 'country name'})
data.head()
```

```python
import matplotlib.pyplot as plt
plt.bar(data['country name'],data['Gold'])
plt.xticks(rotation=45)
plt.show()
```

```python
plt.bar(data['country name'],data['Gold'])
plt.xticks(rotation=30)
plt.title('Countries with most Gold Medals')
plt.xlabel('Country')
plt.ylabel('Gold Medals')
plt.show()
```

Using pandas for visualization

```python
data['Gold'].plot.bar()
```

IQR= Q3-Q1

```python
import matplotlib.pyplot as plt

# Calculate total medals for each country
population = data['Gold'] + data['Silver'] + data['Bronze']
countries = data.index

# Create the pie chart
plt.pie(population, labels=countries, autopct="%1.1f%%")
plt.title('Distribution of Total Medals by Country')
plt.axis('equal') # Ensures the pie chart is circular
plt.show()
```

Defining Class and Creating Objects

```python
class BankAccount():
    def __init__(self, name, balance):
        self.name = name
        self.balance = balance

    def deposit(self, amount):
        self.balance += amount
        print(f"Deposited {amount}\nNew balance is {self.balance}")

    def withdraw(self, amount):
        if amount <= self.balance:
            self.balance -= amount
            print(f"Withdrawn {amount}\nNew balance is {self.balance}")
        else:
            print("Insufficient Funds!!!")

    def __str__(self):
        return f"Account details\nName: {self.name}\nBalance: {self.balance}"

    def __repr__(self):
        return f"BankAccount(name='{self.name}', balance={self.balance})"

# Creating account
acc1 = BankAccount("Rajendran", 1000)
acc2 = BankAccount("Vijai", 2000)

# Using repr
print(repr(acc1))  # Output: BankAccount(name='Rajendran', balance=1000)

# Using str
print(acc1)        # Output: Account details\nName: Rajendran\nBalance: 1000
```

```python

class SavingsAccount(BankAccount):
    def __init__(self, name, balance, int_rate):
        super().__init__(name, balance)
        self.int_rate = int_rate

    def add_interest(self):
        interest = self.balance * self.int_rate
        self.balance += interest
        print(f"Interest added: {interest}\nNew balance: {self.balance}")

    def __str__(self):
        return (f"Savings Account Details\n"
                f"Name: {self.name}\n"
                f"Balance: {self.balance}\n"
                f"Interest Rate: {self.int_rate}")

```

```python
#Abstract class
from abc import ABC, abstractmethod

#  base class
class Animal(ABC):

    @abstractmethod
    def speak(self):
        pass

# Subclass Dog
class Dog(Animal):
    def speak(self):
        print("Bow Bow")

# Subclass Cat
class Cat(Animal):
    def speak(self):
        print("Meow Meow")

d = Dog()
c = Cat()


```

```python
d.speak()
c.speak()
```

```python
#Abstract Base Class
from abc import ABC, abstractmethod

class Library(ABC):
    @abstractmethod
    def show_details(self):
        pass
```

```python
class Book(Library):
    def __init__(self, title, author, isbn):
        self.title = title
        self.author = author
        self.isbn = isbn

    def show_details(self):
        print(f"Book Title: {self.title}\nAuthor: {self.author}\nISBN: {self.isbn}")
```

```python
#Member Class with Encapsulation
class Member(Library):
    def __init__(self, name, member_id):
        self.name = name
        self.member_id = member_id
        self.__books_issued = []

    def issue_book(self, book):
        self.__books_issued.append(book)

    def get_issued_books(self):
        return self.__books_issued

    def show_details(self):
        print(f"Member Name: {self.name}\nMember ID: {self.member_id}")
        print("Books Issued:")
        for book in self.__books_issued:
            print(f" - {book.title}")
```

```python
#Inheritance: StudentMember and FacultyMember
class StudentMember(Member):
    def __init__(self, name, member_id, course):
        super().__init__(name, member_id)
        self.course = course

    def show_details(self):
        super().show_details()
        print(f"Course: {self.course}")

class FacultyMember(Member):
    def __init__(self, name, member_id, department):
        super().__init__(name, member_id)
        self.department = department

    def show_details(self):
        super().show_details()
        print(f"Department: {self.department}")
```

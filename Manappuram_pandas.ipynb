```python
#check for non number values and forces them to be
#converted ti numeric values. still not able to convert, then
# error='coerce' --> will make sure that such values are being updated
# as 'NaN'
df_auto['horsepower'] =pd.to_numeric(df_auto['horsepower'],errors='coerce')
df_auto['horsepower'].unique()
```

```python
mean_hp = df_auto.horsepower.mean()
mean_hp
```

```python
#fill the places where it was 'Nan' with the 'Mean_hp'
df_auto['horsepower'].fillna(mean_hp)

# to confirm the number of null values in the column
df_auto['horsepower'].isnull().sum()
```

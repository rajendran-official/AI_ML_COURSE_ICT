**Checking significance of categorical columns**





```python
#mean profit dispersion
#we check each categorical column and the mean profit value associated with each data in each column
cat_cols = df.select_dtypes(include='object').columns
cat_cols
```

```python
result =[]
for col in cat_cols:
  mean_profit_per_category = df.groupby(col)['Profit'].mean()
  result.append({
      'column':col,
      'No_of_categories':df[col].nunique(),
      'mean_of_mean_profit':mean_profit_per_category.mean(),
      'std_of_mean_profit': mean_profit_per_category.std()
  })

cat_rel = pd.DataFrame(result).sort_values(by='mean_of_mean_profit',ascending=False)
cat_rel
```

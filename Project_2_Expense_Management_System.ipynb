Project 2: Expense Management System

Objective:
Create an application to record daily expenses, categorize them, and generate summary
reports using lists and dictionaries.
Description:
The system should allow users to input expenses, view category-wise totals, and save data
to files for persistence.
Key Requirements:
• Use dictionaries to store category-wise expenses and lists to maintain daily records.

• Implement functions to add_expense(), view_summary(), and export_data().

• Write all expenses to a file `expenses.txt`.

• Include try-except blocks for invalid data entries or file errors.

Suggested File Structure:
• expense_system.py
• expenses.txt
• summary.txt

Evaluation Criteria:
• Data structure usage (25%)
• File handling and report generation (25%)
• Error handling and user validation (25%)
• Team collaboration and code organization (25%)

```python
expenses = []
category_totals = {}

def add_expense():
    try:
        category = input("Enter category (e.g., Food, Travel, Bills): ")
        description = input("Enter description: ")
        amount = int(input("Enter expense amount: "))
        if amount <= 100:
            raise ValueError("Amount must be greater than 100.")

        expense = {"amount": amount, "category": category, "description": description}
        expenses.append(expense)

        category_totals[category] = category_totals.get(category, 0) + amount

        print("Expense added successfully!")
    except ValueError as e:
        print(f"Invalid input: {e}")

def view_summary():
    print("\n--- Expense Summary ---")
    for category, total in category_totals.items():
        print(f"{category}: ₹{total}")

def export_data():
    try:
        with open("expenses.txt", "a") as file:
            for exp in expenses:
                file.write(f"{exp['amount']},{exp['category']},{exp['description']}\n")

        with open("summary.txt", "w") as file:
            for category, total in category_totals.items():
                file.write(f"{category}: ₹{total}\n")

        print("Data exported successfully!")
    except IOError as e:
        print(f"File error: {e}")


while True:
    print("\nExpense Management System")
    print("1. Add Expense")
    print("2. View Summary")
    print("3. Export Data")
    print("4. Exit")

    choice = input("Enter your choice: ")
    if choice == "1":
        add_expense()
    elif choice == "2":
        view_summary()
    elif choice == "3":
        export_data()
    elif choice == "4":
        print("Exiting... Goodbye!")
        break
    else:
        print("Invalid choice. Please try again.")


```

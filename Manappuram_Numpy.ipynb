```python
import numpy as np

# Sample sales and costs arrays
sales = np.array([1000, 1200, 800, 1500, 900])
costs = np.array([700, 900, 600, 1000, 750])

# Calculate profit
profit = sales - costs

# Calculate profit percentage
# To avoid division by zero, we can use np.where or add a small epsilon
profit_percentage = (profit / sales) * 100

print("Sales Array:", sales)
print("Costs Array:", costs)
print("Profit Array:", profit)
print("Profit Percentage Array:", profit_percentage)
```

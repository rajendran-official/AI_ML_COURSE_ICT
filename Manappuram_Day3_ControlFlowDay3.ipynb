Divisible by 4 : the year must be evenly divisble by 4.
Century year exception: if the year is a century year(meaning it ends in "100", eg 1900, 2000), it must be evenly divible by 400 to be a leap year
if it is only divisble by 100 but not by 400, it is not a leap year. non century year: if the year is divisible by 4 but not a century year, it is a leap year

```python
year = int(input("Enter a year: "))
is_leap = False
# Divisible by 4 ?
if year % 4 == 0:
    # Century year ?
    if year % 100 == 0:
        # Century year must also be divisible by 400
        if year % 400 == 0:
            is_leap = True
    else:
        # Non-century year divisible by 4 â†’ leap year
        is_leap = True

if is_leap:
    print(f"{year} is a leap year.")
else:
    print(f"{year} is NOT a leap year.")
```

```python
#1.withdrawal
#2.Deposit
#3.Exit
#balance = 10000
balance = 10000
while True:
    print("\n****** Bank Menu ******")
    print("1. Withdrawal")
    print("2. Deposit")
    print("3. Exit")
    print(f"Current Balance: ${balance:.2f}")

    choice = input("Enter your choice (1-3): ")

    if choice == '1':
        try:
            amount = float(input("Enter amount to withdraw: "))
            if amount <= 0:
                print("Withdrawal amount must be positive.")
            elif amount > balance:
                print("Insufficient funds.")
            else:
                balance -= amount
                print(f"Successfully withdrew ${amount:.2f}.")
        except ValueError:
            print("Invalid input. Please enter a numerical amount.")
    elif choice == '2':
        try:
            amount = float(input("Enter amount to deposit: "))
            if amount <= 0:
                print("Deposit amount must be positive.")
            else:
                balance += amount
                print(f"Successfully deposited ${amount:.2f}.")
        except ValueError:
            print("Invalid input. Please enter a numerical amount.")
    elif choice == '3':
        print("Exiting the banking system. Goodbye!")
        break
    else:
        print("Invalid choice. Please enter 1, 2, or 3.")
```

```python
#Collatz Sequence
#Input number N
#If even --> divide by 2, else mutiply by 3 +1; repeat till 1.
N = int(input("Enter a positive integer: "))
if N <= 0:
    print("Please enter a positive integer.")
else:
    print("Collatz sequence:")
    while N != 1:
        print(N, end="  ")
        if N % 2 == 0:
            N = N // 2
        else:
            N = 3 * N + 1
    print(N)
```

```python
#Find all perfect number within 1 and 100
#perfect -> 6 --> factors?(1,2,3,6)--> 1+2+3 =6 perfect
#8 --> (1,2,4,8) --> 1+2+4 =7 Not perfect

# Function to check if a number is perfect
def is_perfect(n):
    # Find all proper divisors (excluding the number itself)
    divisors = [i for i in range(1, n) if n % i == 0]
    return sum(divisors) == n

# Check numbers from 1 to 100
perfect_numbers = [num for num in range(1, 101) if is_perfect(num)]

print("Perfect numbers between 1 and 100:", perfect_numbers)
```
